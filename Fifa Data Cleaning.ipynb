## Cleaning Dataset

### Drop Columns Not Needed for Project

```python
# Drop columns we won't need
fifa_new = fifa.drop(columns = ['photoUrl', 'playerUrl', 'Loan Date End'])
```

```python
# Drop a majority of columns we won't be needing to clean for this project
fifa_new = fifa_new.drop(fifa_new.columns[19:61], axis = 1) 
```

```python
# Check dataset information again
fifa_new.info()
```

### Change Column Datatypes

```python
# Changing the data type from object to datetime for the 'Joined' column
fifa_new.Joined = pd.to_datetime(fifa_new.Joined)
```

```python
# Extracting day, month, and year components into new columns
fifa_new['Day'] = fifa_new['Joined'].dt.day
fifa_new['Month'] = fifa_new['Joined'].dt.month
fifa_new['Year'] = fifa_new['Joined'].dt.year

# Check everything worked
print(fifa_new[['Day', 'Month', 'Year']])
```

```python
# Change height column to appropriate data type

def height_to_in(height_str):
    height_parts = height_str.strip('\'"').split("'")
    feet = int(height_parts[0])
    if len(height_parts) > 1:
        inches = int(height_parts[1])
    else:
        inches = 0
    total_inches = feet * 12 + inches
    return total_inches

fifa_new['Height'] = fifa_new['Height'].apply(height_to_in)
```

```python
# Now change weight column to appropriate data type
# Converts weight column to numeric and removes lbs unit
fifa_new['Weight'] = fifa_new['Weight'].str.replace('lbs', '').astype(float)
```

```python
# Check Height and Weight column conversions worked
fifa_new
```

```python
# Function to convert values in Value, Wage, and Release Clause Columns
def convert(num_str):
    num_str = num_str.replace('€', '')
    if 'M' in num_str:
        return float(num_str.replace('M', '')) * 1000000
    if 'K' in num_str:
        return float(num_str.replace('K', '')) * 1000
    return float(num_str)
```

```python
# Apply function to columns
fifa_new['Value'] = fifa_new['Value'].apply(convert)
fifa_new['Wage'] = fifa_new['Wage'].apply(convert)
fifa_new['Release Clause'] = fifa_new['Release Clause'].apply(convert)
```

```python
# Change column names to indicate currency and check work
fifa_new.rename(columns={'Value': 'Value(€)', 'Wage': 'Wage(€)', 'Release Clause': 'Release Clause(€)'}, inplace=True)

fifa_new
```

### Separate Team & Contract Column into Two

```python
# Remove starting and trailing new line characters from rows
fifa_new['Team & Contract'] = fifa_new['Team & Contract'].str.strip('\n')  # Using str.strip will remove starting and trailing string values
```

```python
fifa_new
```

```python
# Separate Team and Contracts into separate columns
team_contract_split = fifa_new['Team & Contract'].str.split('\n', expand=True)

# Assign the split values to new columns
fifa_new['Team'] = team_contract_split[0].str.strip()
fifa_new['Contract'] = team_contract_split[1].str.strip()

# Separate Contract column into Start and End columns
contract_date_split = fifa_new['Contract'].str.split('~', expand = True)

# Assign the split values to new columns
fifa_new['Contract Start'] = contract_date_split[0].str.strip()
fifa_new['Contract End'] = contract_date_split[1].str.strip()

# Drop unecessary columns
fifa_new = fifa_new.drop(columns = ['Contract', 'Team & Contract'])
```

```python
# Check work to make sure everything worked
fifa_new
```

### Remove Unecessary Characters

```python
# Strip newline character from Hits column
fifa_new['Hits'] = fifa_new['Hits'].str.strip('\n')
```

```python
# Strip star characters from W/F. SM, and IR column, and convert columns to numerical datatype
fifa_new['W/F'] = fifa_new['W/F'].str.strip('★')
fifa_new['SM'] = fifa_new['SM'].str.strip('★')
fifa_new['IR'] = fifa_new['IR'].str.strip('★')

columns_convert = ['W/F', 'SM', 'IR']
fifa_new[columns_convert] = fifa_new[columns_convert].astype(int)
```
